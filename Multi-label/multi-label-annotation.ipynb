```python
import pandas as pd 
```

```python
df = pd.read_csv('prepared_binary_dataset.csv')
```

```python
dfRG = df[df['Final Label']=='RG']
```

```python
dfRG.columns
```

```python
def extract_annotation(line):
    res=""
    if(line[6]=='RG'):
        if(line[27]==True):
            res += "ANRS "
        if(line[28]==True):
            res += "ANRI "
        if(line[29]==True):
            res += "RNFLDS "
        if(line[30]==True):
            res += "RNFLDI "
        if(line[31]==True):
            res += "BCLVS "
        if(line[32]==True):
            res += "BCLVI "
        if(line[33]==True):
            res += "NVT "
        if(line[34]==True):
            res += "DH "
        if(line[35]==True):
            res += "LD "
        if(line[36]==True):
            res += "LC "
    else:
        if(line[7]+line[17]==2):
            res += "ANRS "
        if(line[8]+line[18]==2):
            res += "ANRI "
        if(line[9]+line[19]==2):
            res += "RNFLDS "
        if(line[10]+line[20]==2):
            res += "RNFLDI "
        if(line[11]+line[21]==2):
            res += "BCLVS "
        if(line[12]+line[22]==2):
            res += "BCLVI "
        if(line[13]+line[23]==2):
            res += "NVT "
        if(line[14]+line[24]==2):
            res += "DH "
        if(line[15]+line[25]==2):
            res += "LD "
        if(line[16]+line[26]==2):
            res += "LC "
    return res[0:-1]
```

```python
dfRG['tags'] = [extract_annotation(x) for x in dfRG.values]
```
